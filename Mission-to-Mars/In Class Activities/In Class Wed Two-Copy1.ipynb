```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://globalvoices.org/page/2/"
browser.visit(url)
```

```python
browser.links.find_by_partial_text('No thanks').click()
```

```python
ArtInfo = []
```

```python
def Info():
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    headers = soup.find_all('h3', class_ = 'post-title')
    dates = soup.find_all('span', class_ = 'datestamp')
    stuff = zip(headers, dates)
    for header, date in stuff:
        title = header.text
        date = date['title']
        date = date.replace("_", "")
        ArtDict = {}
        ArtDict["title"] = title
        ArtDict["date"] = date
        ArtInfo.append(ArtDict)
        print (title, date)
        print (ArtDict)
```

```python
for x in range(5):
    Info()
    browser.links.find_by_partial_text('Older').click()
```
